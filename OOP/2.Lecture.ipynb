# Bank Account

```python
class BankAccount:
    def __init__(self):
        self.credit = 0
        
    def deposit(self,amount):
        self.credit = self.credit + amount
    
    def withdraw(self,amount):
        if self.credit >=amount:
            self.credit -=amount
        else:
            print("Insufficient Funds")
    def print_balance(self):
        return self.credit
```

```python
a1 = BankAccount()
```

```python
a1.deposit(100)
```

```python
a1.deposit(3000)
```

```python
a1.print_balance()
```

```python
a2 = BankAccount()
```

```python
a2.deposit(100)
```

```python
a2.print_balance()
```

```python
a2.withdraw(1000)
```

```python
a1.withdraw(100)
```

```python
a1.print_balance()
```

```python
class Car:
    def __init__(self,speed,color):
        self.__speed = speed
        self.__color = color
        
    def set_speed(self,value):
        self.__speed = value
    
    def get_speed(self):
        return self.__speed
    def set_color(self,color):
        self.__color = color
        
    def get_color(self):
        return self.__color
```

```python
c1 = Car(200,'Red')
```

```python
c1.get_speed()
```

```python
c1.get_color()
```

```python
c1.set_color("Blue")
```

```python
c1.get_color()
```

```python
class Rectangle:
    def __init__(self,height,width):
        self.__height = height
        self.__width = width
        
    def set_height(self,value):
        self.__height = value
        
    def get_height(self):
        return self.__height
    
    def set_width(self,value):
        self.__width = value
        
    def get_width(self):
        return self.__width
    
    def are(self):
        return self.__width * self.__height
```

```python
r1 = Rectangle(100,200)
```

```python
r1.set_height(120)
```

```python
r1.are()
```

```python
r1.set_width(20)
```

```python
r1.are()
```

```python
class Coursera:
    def __init__(self,course,tech):
        self.__course = course
        self.__tech = tech
        
    def course(self):
        return self.__course+" : " + self.__tech
    
    def set_tech(self,value):
        self.__tech = value
        
    def get_tech(self):
        return self.__tech
    
    def set_course(self,value):
        self.__course =value
        
    def get_course(self):
        return self.__course
        
```

```python
c1 = Coursera("Machine Learning with Python","AI")
```

```python
c1.get_course()
```

```python
c1.get_tech()
```

```python
c2 = Coursera("Statistics using R","Data Analytics")
```

```python
c2.get_course()
```
