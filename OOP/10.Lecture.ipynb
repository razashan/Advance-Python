```python
import math
class Circle:
    def __init__(self,radius):
        self.__radius = radius
        
    def set_radius(self,radius):
        self.__radius = radius
    def get_radius(self):
        return self.__radius
    def area(self):
        return math.pi * self.__radius **2
    def __add__(self,object_circle):
        return self.radius + object_circle.radius
    def __lt__(self,object_circle):
        return (self.__radius <object_circle.__radius)
    def __eq__(self,object_circle):
        return (self.__radius == object_circle.__radius)
    def __str__(self):
        return "Circle Area" + str(self.area())
```

```python
c1 = Circle(2)
c2 = Circle(4)
print(c1.area())
```

```python
print(c2.area())
```

```python
print(c1<c2)
```

```python
print(c1==c2)
```
